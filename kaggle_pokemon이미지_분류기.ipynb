```python
img_name = sorted(os.listdir(path))
print('img_name: ', img_name)
img_paths = []
for i in img_name:
  img_paths.append(path + i )

print('img_paths: ', img_paths)
```

```python
df2['filepath'] = img_paths
df2.head()
```

```python
poke_type = []
code = []

for i in range(df2.shape[0]):
  if (df2.iloc[i]['Type1'] == 'Grass') or (df2.iloc[i]['Type2']=='Grass'):
    poke_type.append('Grass')
    code.append(1)
  elif (df2.iloc[i]['Type1'] == 'Water') or (df2.iloc[i]['Type2']=='Water'):
    poke_type.append('Water')
    code.append(2)
  elif (df2.iloc[i]['Type1'] == 'Fire') or (df2.iloc[i]['Type2']=='Fire'):
    poke_type.append('Fire')
    code.append(3)
  else:
    poke_type.append('NaN') # not a number: 데이터 없음으로 많이 쓰임 
    code.append('NaN')
  
print(poke_type)
print(code)
```

```python
df2['Type'] = poke_type
df2['code'] = code
new_df = df2.drop(['Type1', 'Type2', 'Name'], axis= 1) # 열방향(y축)으로 제거 
```

```python
new_df 
```

```python
new_df.head()
```

```python
idx_true = new_df['Type'] != 'NaN' # nan이 아니면 true, nan이면 false 넣는다
print(idx_true)
```

```python
# idx_true 가 ture인 인덱스만 저장 
new_df = new_df[idx_true] # 인덱싱을 통해 NaN이 아닌 부분만 가져오기 
new_df.head()
```

```python
type_ = new_df['Type'].tolist()

plt.figure(figsize = (14, 6))

sns.countplot(x = type_)
```

```python
# 데이터 augmentation 
from tensorflow.keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(
    rescale = 1./255,
    rotation_range = 40, 
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    shear_range = .2,
    zoom_range = 0.2, 
    horizontal_flip = True, 
    fill_mode = 'nearest',
    validation_split = 0.1
)
```

```python
train_generator = train_datagen.flow_from_dataframe(
    new_df,
    x_col = 'filepath', y_col = 'Type',
    target_size = (120, 120),
    color_mode = 'rgba', # rgb +  alpha(불투명도)
    class_mode = 'categorical',
    batch_size = 32 
)
```

```python
validation_generator = train_datagen.flow_from_dataframe(
    new_df,
    x_col = 'filepath', y_col = 'Type',
    target_size = (120, 120),
    color_mode = 'rgba', # rgb +  alpha(불투명도)
    class_mode = 'categorical',
    batch_size = 4,
    shuffle = True,
    seed = 1,
    subet = 'validation' 
)
```

```python
image_sample = train_generator.next()[0]

plt.figure(figsize = (10, 10))

for i in range(9):
  plt.subplot(3, 3, i+1)
  plt.imshow(image_sample[i, :, :])
  plt.axis('off')

plt.show()
```

```python
import tensorflow.keras as ks
```

```python
#model 
model = ks.models.Sequential()
#입력층
model.add(ks.layers.Dense(4, input_shape = (120, 120, 4)))
# hidden layer
model.add(ks.layers.Conv2D(64, (7, 7), activation = 'relu'))
model.add(ks.layers.MaxPooling2D((2, 2)))
model.add(ks.layers.Conv2D(128, (7, 7), activation = 'relu'))
model.add(ks.layers.MaxPooling2D((2, 2)))
model.add(ks.layers.Conv2D(256, (7, 7), activation = 'relu'))
model.add(ks.layers.MaxPooling2D((2, 2)))
model.add(ks.layers.Conv2D(512, (7, 7), activation = 'relu'))
model.add(ks.layers.MaxPooling2D((2, 2)))
model.add(ks.layers.Flatten())
model.add(ks.layers.Dense(1024, activation = 'relu'))
# 출력층
model.add(ks.layers.Dense(3, activation = 'Softmax'))

model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['acc'])
model.summary()
```

```python
model.layers 
```

```python
# callback to stop training 일정 정확도가 도달했을 때
class myCallback(ks.callbacks.Callback):
  def on_epoch_end(self, epoch, logs = {}):
    if ((logs.get('acc')>0.7) and (logs.get('val_acc')>0.7)):
      print("\n reached 70% accuracy so stoppng training ")
      self.model.stop_training = True 
callbacks = myCallback()
```

```python
history = model.fit(train_generator, validation_data = validation_generator, batch_size = 20, epochs = 10, 
                    callbacks = [ks.callbacks.ReduceLROnPlateau(),
                                 callbacks]
                    )
```
