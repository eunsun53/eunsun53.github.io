```python
import librosa
letter_file_path = './samples/%s.mp3' % string_
x, sr = librosa.load(letter_file_path)
print('x ', x)
print('sr', sr) # sampling rate (1초에 몇개 sampling 하는지 )
```

```python
import matplotlib.pyplot as plt
plt.plot(x)
```

```python
import IPython
IPython.display.Audio(data= x, rate = sr)
```

```python
random_factor = 2
octaves = 1 + random.random() * random_factor
new_sound1 = librosa.effects.pitch_shift(x, sr, octaves)
```

```python
IPython.display.Audio(data= new_sound1, rate = sr)
```

```python
resample_sr = sr * 7.7
new_sound2 = librosa.resample(new_sound1, sr, resample_sr)
```

```python
IPython.display.Audio(data= new_sound2, rate = resample_sr)
```
